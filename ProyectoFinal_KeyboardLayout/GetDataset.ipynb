```python
from datasets import load_dataset
from unidecode import unidecode
import numpy as np
import pandas as pd
```

```python
ds = load_dataset("bigcode/the-stack", streaming=True, split="train", data_dir="data/c++")
```

```python
def CountDict(file, dic, arr):
    for c in file:
        if(dic.get(c) is None):
            continue
        arr[dic[c]] +=1



def PairDict(file, dic, arr):
    prev = ' '
    first = True
    for c in file:
        if(dic.get(c) is None):
            continue
        if first == True:
            first = False
            prev = dic[c]
            continue
        th = dic[c]
        arr[prev][th]+=1
        prev = th
```

```python
validChars = 'pPyYfFgGcCrRlLaAoOeEuUiIdDhHtTnNsSqQjJkKxXbBmMwWvVzZ$&[{}(=*)+]!#~%7531902468`;,./@\\:<>?^|-\'_"\t\n '
allChars = dict(zip(validChars,np.arange(len(validChars))))
charCount = np.zeros(len(validChars))
pairCount = np.zeros((len(charCount), len(charCount)))
```

```python
err = 0
for i, f in enumerate(iter(ds)):
    file = f['content']
    if(i%1000 == 0):
        print('It:', i)
    if(i == 200000):
        break
    ff = unidecode(file)
    ff = ff.replace('\r', '\n')
    try:
        CountDict(ff, allChars, charCount)
        PairDict(ff, allChars, pairCount)
    except:
        print('error in file',i)
        err += 1
print('Errors:',err)
```

```python
print(charCount, pairCount)
```

```python
charMap = pd.DataFrame(zip(allChars.values(), charCount),index=allChars.keys(), columns = ['id', 'count'])
charMap.head()
```

```python
pairMap = pd.DataFrame(pairCount, columns=allChars.keys(), index=allChars.keys())
pairMap
```

```python
pairMap.to_csv('PairCount.csv')
charMap.to_csv('CharCount.csv')
```
