```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = plt.imread('BlankLayout.png')
print(img.shape)
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1)
ax.imshow(img)
ax.text(180, 260, '0')
dx = 55
dy = 130
dnext = 135
dright = 60
ax.text(180 + dright, 260 , '0')
```

```python
img = plt.imread('BlankLayout.png')
print(img.shape)
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1)
ax.imshow(img)
it = 0
iniX = 180
iniY = 260

X = iniX
Y = iniY
for i in range(10):
    X2 = X
    Y2 = Y
    for j in range(4):
        ax.text(X2, Y2, str(it))
        it += 1
        #ax.text(X2+dright, Y2, str(it))
        it+=1
        X2 += dx
        Y2 += dy
    X+= dnext

X2 = X
Y2 = Y
for j in range(3):
    ax.text(X2, Y2, str(it))
    it += 1
    #ax.text(X2+dright, Y2, str(it))
    it+=1
    X2 += dx
    Y2 += dy
X+= dnext


X2 = X
Y2 = Y
for j in range(2):
    ax.text(X2, Y2, str(it))
    it += 1
    #ax.text(X2+dright, Y2, str(it))
    it+=1
    X2 += dx
    Y2 += dy
X+= dnext

X2 = X +dx
Y2 = Y + dy
ax.text(X2, Y2, str(it))
it += 1
#ax.text(X2+dright, Y2, str(it))
it+=1

X2 = iniX-dnext
Y2 = iniY
ax.text(X2, Y2, str(it))
it += 1
#ax.text(X2+dright, Y2, str(it))
it+=1


fig.savefig('BaseIndexing.png', dpi= 1000)
```
