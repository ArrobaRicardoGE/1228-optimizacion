```python
def ImageLayout(dic, double=False, save=False, path='Demo.png'):
    img = plt.imread('BlankLayout.png')
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(1,1,1)
    ax.imshow(img)
    it = 0
    iniX = 180
    iniY = 250

    X = iniX
    Y = iniY
    for i in range(10):
        X2 = X
        Y2 = Y
        for j in range(4):
            ax.text(X2, Y2, str(dic[it]))
            it += 1
            if double == True:
                ax.text(X2+dright, Y2, str(dic[it]))
            it+=1
            X2 += dx
            Y2 += dy
        X+= dnext

    X2 = X
    Y2 = Y
    for j in range(3):
        ax.text(X2, Y2, str(dic[it]))
        it += 1
        if double == True:
            ax.text(X2+dright, Y2, str(dic[it]))
        it+=1
        X2 += dx
        Y2 += dy
    X+= dnext


    X2 = X
    Y2 = Y
    for j in range(2):
        ax.text(X2, Y2, str(dic[it]))
        it += 1
        if double == True:
            ax.text(X2+dright, Y2, str(dic[it]))
        it+=1
        X2 += dx
        Y2 += dy
    X+= dnext

    X2 = X +dx
    Y2 = Y + dy
    ax.text(X2, Y2, str(dic[it]))
    it += 1
    if double == True:
        ax.text(X2+dright, Y2, str(dic[it]))
    it+=1

    X2 = iniX-dnext
    Y2 = iniY
    ax.text(X2, Y2, str(dic[it]))
    it += 1
    if double == True:
        ax.text(X2+dright, Y2, str(dic[it]))
    it+=1

    if save == True:
        fig.savefig(path, dpi= 1000)
    
```
