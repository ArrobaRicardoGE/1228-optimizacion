```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datos.csv')
X = df['x'].values
Y = df['y'].values
plt.figure()
plt.plot(X, Y, '.')
plt.show()
```

$r(x) = A \sin(Bx + c\pi)$

$P = [A, B, C]$

```python
def objective_function(P, X, Y):
    A, B, C = P
    Yp = A * np.sin(B * X + C * np.pi)
    return np.sum(np.abs(Y - Yp))

def draw_solution(P, X, Y):
    A, B, C = P
    Yp = A * np.sin(B * X + C * np.pi)
    plt.figure()
    plt.plot(X, Y, '.')
    plt.plot(X, Yp)
    plt.show()

P = (1,1,1)
print(objective_function(P, X, Y))
draw_solution(P, X, Y)
```

# Scipy Differential Evolution

```python
from scipy.optimize import differential_evolution
```

```python
bounds = [[-100, 100], # Rango de A
          [-100, 100], # Rango de B
          [-100, 100]] # Rango de C
```

```python
res = differential_evolution(objective_function, bounds, args=(X, Y))
res
```

```python
P = res.x
print(objective_function(P, X, Y))
draw_solution(P, X, Y)
```
